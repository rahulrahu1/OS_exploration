```python
import os

# List contents of a directory
dir_contents = os.listdir('/kaggle/input/stock-market-india/FullDataCsv')
print("Directory Contents:", dir_contents)

# Check if it's a file or directory
for item in dir_contents:
    item_path = os.path.join('/kaggle/input/stock-market-india/FullDataCsv', item)
    
    if os.path.isfile(item_path):
        print(f"{item} is a file.")
    elif os.path.isdir(item_path):
        print(f"{item} is a directory.")


```

```python
import os

# Define a writable directory path
writable_dir_path = '/kaggle/working/new_folder'

# Create a new directory in the writable location
os.makedirs(writable_dir_path, exist_ok=True)
print(f"New directory created at {writable_dir_path}")

# List contents of the original directory
dir_path = '/kaggle/input/stock-market-india/FullDataCsv'
dir_contents = os.listdir(dir_path)
print("Directory Contents:", dir_contents)

# Check if it's a file or directory
for item in dir_contents:
    item_path = os.path.join(dir_path, item)
    if os.path.isfile(item_path):
        print(f"{item} is a file.")
    elif os.path.isdir(item_path):
        print(f"{item} is a directory.")

# Get the size of each item
for item in dir_contents:
    item_path = os.path.join(dir_path, item)
    size = os.path.getsize(item_path)
    print(f"{item} size: {size} bytes")

# Attempt to rename a file or directory (only if it's writable)
old_name = os.path.join(dir_path, 'old_file.txt')
new_name = os.path.join(dir_path, 'new_file.txt')
if os.path.exists(old_name):
    print(f"Cannot rename {old_name} in a read-only filesystem.")
else:
    print(f"{old_name} does not exist.")
```
